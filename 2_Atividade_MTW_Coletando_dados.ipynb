

Atividade:
1. Usando pandas, acesse o site: https://en.wikipedia.org/wiki/Minnesota
2. pegue a tabela:  United States presidential election results for Minnesota
3. mostre apenas as porcentagem dos ultimos 5 anos do partido Republicano



````python
# Acessando o site mencionado acima
import pandas as pd

# Ler todas as tabelas da página
tables = pd.read_html('https://en.wikipedia.org/wiki/Minnesota')

# Ver quantas tabelas foram encontradas
print(len(tables))

# Ver o conteúdo das tabelas
for i, table in enumerate(tables):
    print(f'Tabela {i}:')
    print(table.head())  # Mostra as primeiras linhas de cada tabela
````

Foram encontradas 27 tabelas no site mencionado e a tabela que tratou sobre os resultados da eleição presidencial, a partir dos votos das pessoas de Minnesota foi o 8º.

````python

# Tabela que contém os resultados das eleições presidenciais
election_results = tables[8]

# Visualização das primeiras linhas da tabela para entender sua estrutura
print(election_results.head(10))

````

````python
# Dataframes e Dataframe da Tabela 8

url = 'https://en.wikipedia.org/wiki/Minnesota'
df_list = pd.read_html(url)  # Isso retorna uma lista de DataFrames

# Tabela 8:
df = df_list[8]
print(df)
print(df.columns) # Aqui determina as colunas existentes no dataframe
````

````python
# Filtro dos cinco anos mais recentes da coluna dos republicanos (%)

# Ordenação pelo ano em ordem decrescente e pego os cinco primeiros
recent_years = df.sort_values(by=[('Year', 'Year')], ascending=False).head(5)

# Seleção das colunas de interesse (Ano e Percentual de republicanos)
result_df = recent_years[[('Year', 'Year'),('Republican',    '%'),
            ]]

# Resetar o índice para evitar problemas de sobreposição
result = result_df.reset_index(drop=True)

# Exibir o resultado
print("\nResultado com os cinco anos mais recentes da coluna Republican:")
print(result)
````
