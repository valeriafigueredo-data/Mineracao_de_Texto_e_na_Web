# **1. Bibliotecas**

```python
# Bibliotecas
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import re
import nltk
from nltk.corpus import stopwords
```

# **2. Pré-Processamento**

```python

# Stop words do NLTK
nltk.download('stopwords')

# Lista de stop words em inglês
stop_words = set(stopwords.words('english'))

def preprocess(text):

    # Substitui quebra de linha por espaço
    text = re.sub(r"\n", " ",text)
    # Remove caracteres especiais
    text = re.sub(r'[^\w\s]', '', text)
    # Converte para minúsculas
    text = text.lower()
    # Remove espaços em branco extras
    text = re.sub(r'\s+', ' ', text)
    # Remove stop words
    text = ' '.join([word for word in text.split() if word not in stop_words])
    text = text.strip()
    return text.strip()

# Pré-processamento dos documentos
processed_documents = [preprocess(doc) for doc in raw_documents[:N]]

# Exibição dos documentos pré-processados
for idx, doc in enumerate(processed_documents):
    print(f"========== Doc {idx} ==========")
    print(doc)
```

```python
# Remoção de tokens não alfabéticos
filtered_tokens = [[word for word in tokens if word.isalpha()] for tokens in tokenized_documents]

# Exibição dos tokens filtrados
for idx, tokens in enumerate(filtered_tokens):
    print(f"========== Filtered Tokens for Doc {idx} ==========")
    print(tokens)
```

# **3. TF**

```python
import pandas as pd
import numpy as np

# Criação de um conjunto de todas as palavras únicas
unique_tokens = set(token for tokens in filtered_tokens for token in tokens)

# Criação de um DataFrame para armazenar contagens e TF
df = pd.DataFrame(data=np.zeros((2, len(unique_tokens))), columns=list(unique_tokens), index=["N", "TF"])

# Conta a frequência de cada token
for tokens in filtered_tokens:
    for word in tokens:
        if word in df.columns:
            df.at["N", word] += 1

# Calcula TF (Term Frequency)
total_tokens = df.loc["N"].sum()  # Total de todas as contagens
for token in df.columns:
    df.at["TF", token] = df.at["N", token] / total_tokens if total_tokens > 0 else 0

# Exibi o DataFrame com as contagens e TF
print(df)
```

```python
# Transpor o DataFrame
transposed_df = df.T  # As linhas se tornam colunas e vice-versa
transposed_df
```

```python
# Ordenamento do DataFrame transposto por TF em ordem decrescente
sorted_df = transposed_df.sort_values(by='TF', ascending=False)
sorted_df
```

```python
# 5 Tokens com maior TF
# Encontrar os 5 termos com o maior TF
top_terms = df.loc["TF"].nlargest(5)

# Exibi os 5 termos com maior TF
print("Top 5 termos com maior TF:")
print(top_terms)
```

# **4. IDF**

```python
import math

# Calcular o número total de documentos
total_documents = len(filtered_tokens)

# Criar um dicionário para armazenar o IDF
idf_values = {}

# Calcular IDF para cada token
for token in df.columns:
    # n_t é o número de documentos que contêm o token
    n_t = df.at["N", token]  # Aqui está o número total de ocorrências do token

    # Calcular IDF
    idf_values[token] = math.log10(total_documents / n_t) if n_t > 0 else 0

# Adicionar os valores de IDF ao DataFrame
idf_df = pd.DataFrame.from_dict(idf_values, orient='index', columns=['IDF'])
idf_df.index.name = 'Word'

print("\nDataFrame de IDF:")
print(idf_df)
```

```python
# Ordenamento do DataFrame transposto por IDF em ordem decrescente
sorted_df2 = idf_df.sort_values(by='IDF', ascending=False)
sorted_df2
```

```python
# 5 Tokens com maior IDF
top_5_idf = idf_df.nlargest(5, 'IDF')

# Exibir os 5 termos com maior IDF
print("\nTop 5 termos com maior IDF:")
print(top_5_idf)

```

# **5. TF-IDF**

```python

# Criar um DataFrame para armazenar o TF-IDF
tf_idf_df = pd.DataFrame(index=df.columns, columns=['TF-IDF'])

# Calcular o TF-IDF
for token in df.columns:
    tf_idf_df.at[token, 'TF-IDF'] = df.at["TF", token] * idf_df.at[token, 'IDF']

# Exibir o DataFrame com os resultados de TF-IDF
print(tf_idf_df)

```

# **6. Dataframe com todos Indicadores**

```python
# Exibição do DataFrame com os resultados de TF, IDF e TF-IDF

# Combinar os DataFrames em um único DataFrame
combined_df = pd.concat([df.loc["TF"], idf_df['IDF'], tf_idf_df['TF-IDF']], axis=1)
combined_df.columns = ['TF', 'IDF', 'TF-IDF']

# Ordenar o DataFrame pelo valor de TF-IDF do maior para o menor
combined_df = combined_df.sort_values(by='TF-IDF', ascending=False)

# Exibir o DataFrame combinado e ordenado
print("DataFrame Combinado de TF, IDF e TF-IDF (ordenado pelo TF-IDF):")
print(combined_df)
```

```python
# Converter a coluna 'TF-IDF' para float
tf_idf_df['TF-IDF'] = tf_idf_df['TF-IDF'].astype(float)

# 5 Tokens com maior TF-IDF
top_5_tfidf = tf_idf_df.nlargest(5, 'TF-IDF')

# Exibir os 5 termos com maior TF-IDF
print("Top 5 termos com maior TF-IDF:")
print(top_5_tfidf)
```

# **7. Conclusão**



No presente trabalho foi elaborado um DataFrame com o TF, IDF e TF-IDF de todos os termos (tokens). Para isso foi realizada a limpeza dos dados, o seu pré-processamento para eliminar sujeiras, caracteres, espaços extras, enfim, dados não essenciais para a análise.

O TF (Frequência do Termo) significa a frequência de um termo em um documento específico. Quanto mais vezes um termo aparece em um documento, mais importante ele é considerado nesse documento.

O IDF (Frequência Inversa de Documento) quantifica a importância de um termo em relação a um conjunto de documentos (ou corpus).Isto porque termos que aparecem frequentemente em muitos documentos não são tão informativos como "e", "ou", "mas", de modo que não agregam valor significativo à análise.

O TF-IDF, por sua vez, significa Frequência do Termo - Frequência Inversa do Documento. É uma medida usada para avaliar a importância de uma palavra em um documento dentro de um conjunto de documentos (ou corpus). Ele combina as duas informações: TF e IDF. Logo, são termos que ocorrem frequentemente em um documento mas de forma rara nos demais.

Sobre os valores da medida TF-IDF, não tem um limite fixo, variam entre valores negativos, 0 a valores superiores a 1, dependendo da frequência do termo e de sua distribuição nos documentos. TF- IDF = 0 significa que o termo não aparece no documento ou que é uma palavra muito comum que aparece em todos os documentos, resultando em um IDF baixo. E quanto mais afastado de 0, positivo, indica que o termo é relevante para o documento e ao corpus como um todo.

O Top 5 termos com maior TF foi would, organization, subject, lines e im. Isso porque tais termos têm alta frequência nos documentos. Por exemplo, subject, lines  e organization estão na parte do cabeçalho dos e-mails e todos textos são emails.

O Top 5 termos com maior TF-IDF foram destruction, performance, use, station e board  (0.000958). Isso significa que apareceram bastante em um documento (texto) e pouco nos demais. POr exemplo, o termo destruction apareceu bastante no texto 6 (DOC 5), abordando sobre armas de destruição em massa e o seu impacto na sociedade. Entretanto, para os demais textos (DOC) tal termo apareceu de forma irrelevante.
